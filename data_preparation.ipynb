```python
import cv2 
from mtcnn import MTCNN
import os
```

```python
def detect_face(img):
    detector = MTCNN()
    faces = detector.detect_faces(img)
    return faces[0]['box'] if len(faces) > 0 else None
```

```python

def crop_face(img, box):
    x, y, w, h = box
    return img[y:y+h, x:x+w]
```

```python
path = 'C:/Users/Ponda/OneDrive/desktop/Linear/RawData/Test/'
```

```python
# crop image and save to train folder
def crop_dataset(path, new_path):
    print(path)
    for filename in os.listdir(path):
        # print(path+filename)
        img = cv2.imread(path + filename)
        box = detect_face(img)
        if box is not None:
            face = crop_face(img, box)
            if not os.path.exists(new_path):
                os.makedirs(new_path)
            cv2.imwrite(new_path+filename, face)
```

```python
classes_path_list = ['Blue', 'New', 'Pond', 'Tar']
new_path = 'C:/Users/Ponda/OneDrive/desktop/Linear/Test/Images/'
```

```python
for pth in classes_path_list:
    crop_dataset(path+'/'+pth+'/', new_path+pth+'/')
```

```python
# to rename files in a folder to a specific name
def rename_files(path, prefix):
    i = 0
    
    for filename in os.listdir(path):
        dst = prefix + str(i) + ".jpg"
        src = path + filename
        dst = path + dst
        
        os.rename(src, dst)
        i += 1

rename_files(path+'New/', 'New-')
print(path+'New/')
```
